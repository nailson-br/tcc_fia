```python
import pandas as pd
from pandas import json_normalize
import json
```

```python
# Caminho para o arquivo JSON
json_file = "D:/repositorios/tcc_fia/sources/municipios.json"
```

```python
print(json_file)
```

```python
# Carrega o JSON e cria um DataFrame Pandas
with open(json_file, encoding='utf-8') as f:
    json_data = json.load(f)
```

```python
# Flatten do JSON e criação do DataFrame Pandas
df = json_normalize(json_data)
```

```python
# Mostra o DataFrame resultante
print(df)
```
